# Luxor
Vector graphics

```julia
using Luxor
```

```julia
function make_drawing(r, c, style)
    # Draw in SVG format, storing directly in memory
    d = Drawing(300,300,:svg)
    origin(150,150)
    sethue(c)
    circle(Point(0,0), r, style)
    finish()
    return d
end
```

```julia
make_drawing(100,"red",:fill)
```

```julia
function make_gui()
    # Observable drawing
    drawing = Observable(make_drawing(100, "black", :stroke))
    # some controls
    colors = radiobuttons(["red", "green", "blue"])
    style = dropdown([:fill, :stroke])
    radius = slider(10:150)
    # Update function
    update(::Any) = (drawing[] = make_drawing(radius[], colors[], style[]))
    # GUI layout
    gui = vbox(
        hbox(colors, style),
        radius,
        drawing
    )
    # Register update functions
    on(update,colors)
    on(update,radius)
    on(update,style)
    # Update drawing with initial values
    update(0)
    return gui
end
```

```julia
make_gui()
```

## A more complicated example

```julia
using Unitful
using GR
```

```julia
function plot_compression(ratios, n)
    Ta = 296u"K"
    pa = 101325u"Pa"
    Cp = 1004u"J/(kg*K)"
    r = 287.11u"J/(kg*K)"
     = Cp/(Cp-r)
    
    # Entropy range
    srange = (-1500:100).*(u"J/(kg*K)")
    
    # Temperature as function of entropy for the compression
    T(s,s0) = Ta*exp((s-s0)/(Cp - r*n/(n-1)))
    
    # Temperature as function of entropy for the isobars
    isobar(T,s) = T*exp(s/Cp)

    # Isothermal at Ta
    plt = plot(ustrip.(srange), fill(ustrip(Ta),size(srange)))

    # Plot the initial isobar
    plt = oplot(ustrip.(srange), ustrip.(isobar.(Ta,srange)))
    
    Tis = Ta
    si = 0.0u"J/(kg*K)"
    # Plot the other isobars
    for c in ratios
        Tis = Tis*c^((-1)/) # Temperature on the isobar at s = 0
        plt = oplot(ustrip.(srange), ustrip.(isobar.(Tis,srange)))
        # Polytropic compression temperature range
        Tend = Ta*c^((n-1)/n)
        s0 = si
        si = s0 + Cp * log(Tend / Ta) - r * log(c)
        s = range(s0,stop=si,length=100)
        plt = oplot(ustrip.(s), ustrip.(T.(s,s0)))
        si -= Cp*log(Tend/Ta)
    end
    return plt
end
```

```julia
plot_compression([5,5,5],1.2)
```

```julia
# Build a GUI to visualize and modify the compression stages, for the given ratios and polytropic coefficient
function compression_ui(ratios,n)
    mdot = 10u"kg/s"
    Ta = 296u"K"
    pa = 101325u"Pa"
    Cp = 1004u"J/(kg*K)"
    r = 287.11u"J/(kg*K)"
     = Cp/(Cp-r)
    Pc(c,n) = mdot*n/(n-1)*r*Ta*(c^((n-1)/n)-1)
    P = sum(Pc.(ratios,n))
    global_ratio = prod(ratios)
    sliders = []
    output_plot = Observable(plot_compression(ratios,n))
    current_power = Observable(uconvert(u"kW", P))
    minimum_power = Observable(uconvert(u"kW", P))
    for i in 1:length(ratios)-1
        s = slider(range(0.1,stop=ratios[i]*2, length=100); value=ratios[i])
        on(s) do c
            ratios[i] = c
            ratios[end] = global_ratio/prod(ratios[1:end-1])
            current_power[] = sum(Pc.(ratios,n))
            minimum_power[] = min(minimum_power[], current_power[])
            output_plot[] = plot_compression(ratios, n)
        end
        push!(sliders, s)
    end
    return vbox(
        sliders...,
        output_plot,
        hbox("Minimum power: ", minimum_power),
        hbox("Current power: ", current_power)
    )
end
```

```julia
compression_ui([5.0,5.0,5.0],1.2)
```

```julia
using Unitful: 
function labeled_arrow(vax, vu, label, opt=O, flip=false)
    scaling = 100
    endpt = scaling*Point(vax,-vu)
    if flip
        arrow(opt-endpt, opt)
        Luxor.text(label, opt-endpt-(10,-10), halign=:left)
    else
        arrow(opt,opt+endpt)
        Luxor.text(label, opt+endpt+(10,-5), halign=:left)
    end
    return endpt
end

function plot_blade(startpos,vin_unflipped,vout_unflipped) 
    s = 50 # scaling factor
    cord = 100
    
    vin = Point(vin_unflipped.x, -vin_unflipped.y)
    vout = Point(vout_unflipped.x, -vout_unflipped.y)
    
    endpos = Point(cord+startpos.x, 1.5*s*(vout.y+vin.y)+startpos.y)
    
    pts = [startpos+s*vin/sqrt(dotproduct(vin,vin)), endpos-s*vout/sqrt(dotproduct(vout, vout)), endpos]
        
    sethue("black")
    move(startpos)
    curve(pts...)
    strokepath()
    return (startpos, endpos)
end

function plot_blades(1,2,3)
     = 0.6
    u = -1.0
    
    # Velocity calculation
    v0u = 0
    v1u = *tan(1)
    w1u = v1u - u
    w2u = *tan(2)
    v2u = w2u + u
    v3u = *tan(3)
    
    d = Drawing(700,500,:svg)
    origin(100,250) # origin at center-left of drawing
    
    # Draw the main blade row
    setline(3.0)
    (igvin, igvout) = plot_blade(O,Point(,v0u), Point(,v1u))
    (rotin, rotout) = plot_blade(Point(160,0),Point(,w1u), Point(,w2u))
    (statin, statout) = plot_blade(Point(320,0), Point(,v2u), Point(,v3u))
    
    # One extra row above and below
    setopacity(0.4)
    plot_blade(Point(0,80),Point(,v0u), Point(,v1u))
    plot_blade(Point(160,80),Point(,w1u), Point(,w2u))
    plot_blade(Point(320,80), Point(,v2u), Point(,v3u))
    plot_blade(Point(0,-80),Point(,v0u), Point(,v1u))
    plot_blade(Point(160,-80),Point(,w1u), Point(,w2u))
    plot_blade(Point(320,-80), Point(,v2u), Point(,v3u))
    setopacity(1.0)
    setline(1.0)

    # Velocity vectors at the blades
    v0 = labeled_arrow(, v0u, "v0", igvin, true)
    
    sethue("blue")
    v1 = labeled_arrow(, v1u, "v1", igvout, false)
    v2 = labeled_arrow(, v2u, "v2", statin, true)
    v3 = labeled_arrow(, v3u, "v3", statout, false)
    
    sethue("red")
    w1 = labeled_arrow(, w1u, "w1", rotin, true)
    w2 = labeled_arrow(, w2u, "w2", rotout, false)
    
    # Velocity triangle at origin
    P0 = Point(-100*, 0)
    sethue("blue")
    labeled_arrow(, v2u, "v2", P0, false)
    labeled_arrow(, v3u, "v1=v3", P0, false)
    
    sethue("red")
    labeled_arrow(, w1u, "w1", P0, false)
    labeled_arrow(, w2u, "w2", P0, false)
    
    # Degree of reaction
    r = (w1u+w2u)/(-2u)
    
    sethue("black")
    Luxor.text("r = $(round(r;digits=5))", Point(210,-230), halign=:center)
    Luxor.text("IGV", Point(50,-200), halign=:center)
    Luxor.text("Rotor", Point(210,-200), halign=:center)
    Luxor.text("Stator", Point(370,-200), halign=:center)

    v0 = 
    v1 = sqrt(^2+v1u^2)
    v2 = sqrt(^2+v2u^2)
    v3 = sqrt(^2+v3u^2)
    
    # Enthalpy
    dh_igv = round((v0^2-v1^2)/2;digits=5)
    dh_rot = round(u*(v2u-v1u) - (v2^2-v1^2)/2;digits=5)
    dh_stat = round((v2^2-v3^2)/2;digits=5)
    
    Luxor.text("h = $dh_igv", Point(50,-180), halign=:center)
    Luxor.text("h = $dh_rot", Point(210,-180), halign=:center)
    Luxor.text("h = $dh_stat", Point(370,-180), halign=:center)
    r_checked = round(dh_rot/(dh_rot + dh_stat);digits=5)
    Luxor.text("Check r: $r_checked", Point(210,-160), halign=:center)
    
    finish()
    return d
end
```
